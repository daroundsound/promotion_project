```python
import openpyxl
import requests
from openpyxl.drawing.image import Image
from io import BytesIO

# 엑셀 파일 열기
wb = openpyxl.load_workbook("C:\Users\NSE\Downloads\키워드를 위한.csv")
ws = wb.active

# 이미지가 들어갈 열과 행 범위를 지정합니다.
# 예: B열에 이미지 링크가 있는 경우
image_column = 'E'
new_image_column = 'N'
start_row = 2  # 시작 행 (헤더가 있는 경우)
end_row = ws.max_row  # 데이터가 있는 모든 행을 처리

for row in range(start_row, end_row + 1):
    image_url = ws[f'{image_column}{row}'].value
    if image_url:
        try:
            # 이미지 다운로드
            response = requests.get(image_url)
            if response.status_code == 200:
                img_data = BytesIO(response.content)
                img = Image(img_data)
                
                # 이미지 크기 조정 (선택 사항)
                img.width = 100
                img.height = 100

                # 이미지 삽입
                ws.add_image(img, f'{new_image_column}{row}')

        except Exception as e:
            print(f'Failed to load image for {image_url}: {e}')

# 엑셀 파일 저장
wb.save('C:\\Users\\NSE\\Downloads\\Advertising_combined_all_rows.xlsx')

```
